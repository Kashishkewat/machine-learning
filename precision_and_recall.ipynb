```python
df=pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
y_pred1=clf1.predict(X_test)
y_pred2=clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('logistic regression',accuracy_score(y_test,y_pred1))
print("decision tree ",accuracy_score(y_test,y_pred2))
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
from sklearn.metrics import recall_score,precision_score,f1_score
```

```python
print("for logistic regression model")
print("-"*50)
cdf=pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,2)))
print(cdf)
print("-"*50)
print("precision_",precision_score(y_test,y_pred1))
print("recall_",recall_score(y_test,y_pred1))
print("f1_score_",f1_score(y_test,y_pred1))
```

```python
print("for decision tree model")
print("-"*50)
cdf=pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,2)))
print(cdf)
print("-"*50)
print("precision_",precision_score(y_test,y_pred2))
print("recall_",recall_score(y_test,y_pred2))
print("f1_score_",f1_score(y_test,y_pred2))
```
